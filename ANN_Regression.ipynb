```python
import pandas as pd
# Mengimpor dataset
dataset = pd.read_csv('Cancer_Data.csv')
dataset.head()
```

```python
dataset.info()
```

```python
# Mendrop kolom yang tidak diperlukan
dataset.drop(columns="Unnamed: 32", inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder

# Melakukan encoding pada kolom yang bertipe data kategorikal

# Membuat objek LabelEncoder
label_encoder = LabelEncoder()

# Iterasi pada setiap kolom di DataFrame
for column in dataset.columns:
    # Mengecek apakah ktipe data pada kolom adalah object (Categorical)
    if dataset[column].dtype == 'object':
        # Melakukan encoding pada kolomnya
        dataset[column] = label_encoder.fit_transform(dataset[column])

dataset
     
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Mencari korelasi dari fitur - fitur yang akan divisualisasikan

# Pilih kolom yang diperlukan
selected_columns = ['diagnosis', 'radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean',
                    'smoothness_mean', 'compactness_mean', 'concavity_mean', 'concave points_mean']

# Buat subset dataset dengan kolom yang dipilih
selected_data = dataset[selected_columns]

# Hitung matriks korelasi
correlation_matrix = selected_data.corr()

# Visualisasikan matriks korelasi sebagai heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
# Hapus spasi ekstra dari nama kolom
dataset.columns = dataset.columns.str.strip()

# Pilih kolom yang diperlukan
selected_columns = ['diagnosis', 'radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean','smoothness_mean', 'compactness_mean', 'concavity_mean', 'concave points_mean']

# Buat pair plot untuk visualisasi setiap pasang fiturnya
sns.pairplot(dataset[selected_columns], hue='diagnosis', palette='coolwarm', diag_kind='kde')
plt.show()
```

```python
#Menentukan x dan y untuk dipelajari model
X = dataset.iloc[:, [4]].values
y = dataset.iloc[:, 2].values
```

```python
# Membagi data ke test dan training set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Mengimpor library Keras dan turunannya
# Sekarang menggunakan TensorFlow 2.0
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Inisialisasi ANN
MesinRegresi = Sequential()

# Menambah input layer dan hidden layer pertama
MesinRegresi.add(Dense(units=10, kernel_initializer='uniform', activation='relu'))

# Menambah hidden layer kedua
MesinRegresi.add(Dense(units=8, kernel_initializer='uniform', activation='relu'))

# Menambah output layer
MesinRegresi.add(Dense(units=1, kernel_initializer='uniform', activation='linear'))

# Meenjalankan ANN
MesinRegresi.compile(optimizer='adam', loss='mean_squared_error')

# Fitting ANN ke training set
MesinRegresi.fit(X_train, y_train, batch_size=10, epochs=100)

# Predict on test data
y_pred = MesinRegresi.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

# Calculate RMSE
rmse_modified = np.sqrt(mean_squared_error(y_test, y_pred))

# Calculate MAE
mae_modified = mean_absolute_error(y_test, y_pred)

# Calculate R^2
r_squared = r2_score(y_test, y_pred)

# Print modified metrics
print("Root Mean Square Error (RMSE):", rmse_modified)
print("Mean Absolute Error (MAE):", mae_modified)
print("R-squared (R^2):", r_squared)
```

```python
import matplotlib.pyplot as plt

# Scatter plot for actual vs. predicted values
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, color='blue')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red') 
plt.title('Visualization')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.show()
```
